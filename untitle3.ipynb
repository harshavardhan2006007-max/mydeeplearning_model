```python
from sklearn.datasets import load_digits
```

```python
x,y=load_digits(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_pre,x_test,y_pre,y_test=train_test_split(x,y,test_size=0.2,train_size=100)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_pre=scaler.fit_transform(x_pre)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model1=LogisticRegression(max_iter=1000)
model1.fit(x_pre,y_pre)
train=model1.predict(x_test)
accuracy_score(train,y_test)
```

```python
model2=DecisionTreeRegressor(max_depth=100000)
model2.fit(x_pre,y_pre)
train=model2.predict(x_test)
accuracy_score(train,y_test)
```

```python
model3=RandomForestRegressor(n_estimators=100)
model3.fit(x_pre,y_pre)
train6=model3.predict(x_test)
```

```python
accuracy_score(train6,y_test)
```
