1.Numpy import

```python
import numpy as np
```

```python
x=np.random.rand(10,10)
```

```python
x.shape
```

```python
x.dtype
```

```python
x.size
```

```python
x
```

```python
b=np.random.rand(10,10)
```

```python
b
```

```python
w=np.random.rand(10,10)
```

```python
w
```

```python
y=2*x+2 #True Y
```

```python
y
```

Training Loop

```python
epoach=100
learning_rate=0.2
for epoach in range(epoach):
    y_hat=x@w+b
    loss=((y_hat-y)**2).mean()
    w-=learning_rate*loss
    b-=learning_rate*loss
    print(loss)
```

Testing

```python
y_test=np.random.rand(10,10)
loss=((y_hat-y_test)**2).mean()
print(loss)
```
