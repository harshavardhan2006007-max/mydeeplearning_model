###linearDeepmodel start
*input and real output

```python
import torch
import torch.nn as nn
torch.manual_seed(100)
X=torch.randn(2,2)
Y=torch.randn(2,2)
```

##Model

```python
class model(nn.Module):
  def __init__(self):
    super().__init__()
    # Layer 1: Input (1) -> Hidden (2)
    self.linear1 = nn.Linear(2, 2)
    self.hidden_act = nn.SELU()
    # Layer 2: Hidden (2) -> Output (1)
    self.output_linear = nn.Linear(2, 2)

  def forward(self, x):
    # Pass through the first layer
    x = self.linear1(x)
    # Apply activation
    x = self.hidden_act(x)
    # Pass through the output layer
    x = self.output_linear(x)
    return x
```

###Training Loop

```python
Model=model()
learning_rate=0.2
```

```python
epoach=100
optimer=torch.optim.SGD(Model.parameters(),learning_rate)
critian=nn.MSELoss()
```

```python
from typing import ItemsView
for epoah in range(epoach):
  Y_hat=Model(X)
  loss=critian(Y_hat,Y)
  optimer.zero_grad()
  loss.backward()
  optimer.step()
  print(Model.state_dict())
```

#Manual InearRegression

```python
import torch
X=torch.randn(10,10)
Y=torch.randn(10,10)
learning_rate=0.1
```

###weights&Bias

```python
w=torch.randn(10,10,requires_grad=True)
b=torch.randn(10,10,requires_grad=True)
```

####Training Loop

```python
from os import stat_result
from numpy import mean
epoach=1000
for epoch in range(epoach):
  Y_hat=X@w+b
  Loss=((Y_hat-Y)**2).mean()
  Loss.backward()
  with torch.no_grad():
    w-=learning_rate*w.grad
    b-=learning_rate*b.grad
    # w.grad.zero_()
    # b.grad.zero_()
    if epoch % 10 == 0:
      print(f'Epoch {epoch}, Loss: {Loss.item()}')
    w.grad.zero_()
    b.grad.zero_()
```
